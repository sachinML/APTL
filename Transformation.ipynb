```python
from pyspark.sql.functions import col, udf
from pyspark.sql.types import DoubleType
import string

# Define the function that converts a string to a double
def string_to_double(x):
    if "," in x:
        translator = str.maketrans('', '', string.punctuation)
        x = x.translate(translator)
    
    return float(x)
```

```python
# Convert the function to a UDF
string_to_double_udf = udf(string_to_double, DoubleType())

# Use withColumn to apply the UDF to the column
df4 = df3.withColumn("price", string_to_double_udf(col("price")))
```

```python
df4.show()
```

```python
df4.printSchema()
```

```python
df5 = df4.withColumn("star", string_to_double_udf(col("star")))
```

```python
df5.printSchema()
```
